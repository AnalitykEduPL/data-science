# Rozkład normalny w Python

## Generowanie zmiennej o rozkładzie zblizonym do normalnego

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
normalVar = pd.DataFrame(np.random.normal(0,10,70))
normalVar.columns = ['value']
normalVar.head()
```

```python
normalVar.size
```

# Weryfikacja graficzna

## Histogram

```python
sns.distplot(normalVar)
plt.show()
```

## Q-Q Plot

```python
import seaborn_qqplot as sqp
from scipy.stats import gamma
```

```python
sqp.qqplot(normalVar, 
           x='value',
           y=gamma,
          aspect=2.5,
           height = 4,
           display_kws={"identity":True}
          )
plt.title("QQ Plot")
plt.show()
```

# Testy statystyczne

## Shapiro-Wilk Test

```python
from scipy.stats import shapiro
stats, p = shapiro(normalVar)
print(stats,p)
if p > 0.05:
    print ("Rozkład wygląda na normalny")
```

## D'Agostino-Pearson

```python
from scipy.stats import normaltest
stats, p = normaltest(normalVar)
print(stats, p)
if p > 0.05:
    print ("Rozkład wygląda na normalny")
```
